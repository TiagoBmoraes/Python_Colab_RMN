No Colab podemos aproveitar todo o potencial das conhecidas bibliotecas Python para simular, analisar e ver dados. A célula de códigos abaixo usa **numpy** e **matplotlib**. Para editar o código, basta clicar na célula e começar a editar.

Vamos começar explorando uma equação do tipo: 
$$ S(t) = A sen(\omega t) $$

```python
import numpy as np
from matplotlib import pyplot as plt

# Linhas Verdes com # são comentários e não são executados nos códigos
# Vamos começar criando um sinal do tipo S(t) = A * sen (wt);
# Amplitude do Sinal (u.a.):
A = 5

#  Frequência de oscilação Omega w (rad/s)
Omega = 2

# Eixo do tempo (começando em 0 e indo até 10 segundos, com 1000 pontos)
tempo = np.linspace(0, 10, 1000)

# Equação S(t) = A * sen (w*t)
Sinal = A * np.sin(Omega*tempo)

# Códigos abaixo é o que geram a Figura:
plt.plot(tempo, Sinal, '-', label=r"$S(t) = A * sen (w*t)$")
plt.fill_between(tempo, Sinal, 195, where=(Sinal > 195), facecolor='g', alpha=0.6)
plt.grid(True, linestyle='-.')
plt.legend(loc=1) # upper rigth corner
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (u.a.)')
plt.title("Sinal oscilatório: S(t) = A * sen (w*t)")
plt.show()
```

# Precessão do Spin Nuclear

Para fazer **RMN** precisamos aplicar um Campo Mangético estátio na nossa amostra, então seja $$\vec{B}_{0}\hat{z}$$ nosso campo que aponta na direção z.

RMNequipamento.png

Quando estão imersos em um Campo Magnético estátio 
$$\vec{B}_{0}\hat{z}$$, apontando na direação $$\hat{z}$$ por exemplo, os spins nucleares precessionam ao redor da direção do campo mangético externo

$$\Omega_{0} = \gamma B_{0} $$

```python
# Intensidade do Campo Magnético B0 (Tesla)
# Sugestão, veja a Frequencia de Larmos para campos de B0 =  1 | 9.4 | 14.1 Teslas
B0 = 1

# Escolha o Núcleo [ 1H, 13C, 15N, 23Na, 29Si, etc..  ] 
# 1H gyromagnetic ratio 42.58  [ MHz / Tesla ] 
gamma_1H = 42.58

# Precessão Larmor ( W0 = gama * B0 )  (em Hz)
Precessao =  gamma_1H * B0 
print(Precessao,'MHz')
```

O Spin Nuclear do 1H é **I = 1/2**



```python
 # importanto: Physical Constants
 import scipy.constants as sc

# Spin Nuclear (I)
# Momento angular ( L = hcortado * I)
# Momento magnético nuclear ( mi = gama* L = gama * hcortado * I)
I = 1/2

# Momento magnético nuclear   (hbar no scipy é a constante de Planck: hcortado)
mi = gamma_1H * sc.hbar * I;

print(mi,'verificar unidades...')
```

# Efeito Zeeman

Devido ao Efeito Zeeman, na presença do campo magnético os spins nucleares se desdobram em dois níveis de energia a e b, com energias:



```python
# Níveis de Energia
# E_alpha = + h * gama * B0
# E_beta  = - h * gama * B0

E_alpha = + sc.hbar * gamma_1H * B0;
E_beta  = - sc.hbar * gamma_1H * B0;

# Diferença de Energia entre nível alpha e beta
Delta_E = E_alpha - E_beta;

print(Delta_E,'verificar unidades...')
```

Dessa forma, vemos que a energia necessária para promover um Spin do level beta para o alpha é igual:



```python
# Energia necessária para Transição
# Delta_E = hcortado * W0
frequencia_larmor = Delta_E / (2*np.pi*sc.hbar);

print(frequencia_larmor,' Mhz !')

print('ou seja, devemos pulsar na frequência de =',frequencia_larmor ,'MHz !')
```

# População de spins: Distribuição de Bolztmann

Os níveis de energia não apresentam a mesma quantidade de população de spins, a população é dada pela Distribuição de Bolztmann:



```python
# Distribuição de Boltzmann depende da Temperatura (kelvin)
Temperatura = 300 

# População N_alpha e População N_beta

Razao_N_beta_N_alpha = np.exp(-Delta_E/sc.k * Temperatura)

print('N_beta / N_alpha = ', Razao_N_beta_N_alpha)

# em 1 mol de H2O [ 6,022 * 10^23 ] a diferença de população é:
N_beta = (Razao_N_beta_N_alpha * 6.02 * 10**23) / (1 + Razao_N_beta_N_alpha);
N_alpha =  sc.N_A - N_beta;

Diferenca = - N_beta + N_alpha

print('ou seja,', Diferenca, 'núcleos de diferença')
```

# Pulsos e Relaxação 


A sequência de pulso mais simples possível é a aplicação de  um único pulso de ângulo $\theta$. Após o pulso um sinal FID (*Free Induction Decay*) é detectado, como mostra a figura:

image.png


O sinal FID da figura decai com uma constante de tempo denominada $T_{2}^{*}$, que depende da inomogeneidade do campo magnético aplicado na amostra.


```python
# Decaimento T22 = T2*; 
T22 = 2;   

# Amplitude; Frequência;
A = 5 ;    Omega = 10;

# Eixo do tempo 
tempo = np.linspace(0, 10, 1000)

# Equação S(t) = A * sen (w*t) * exp(-t/T22)
FID_Real = A * np.cos(Omega*tempo) * np.exp(-tempo/T22) 
FID_Imag = A * np.sin(Omega*tempo) * np.exp(-tempo/T22)

# Módulo
Modulo = ( FID_Real**2 + FID_Imag**2 )**0.5


# Códigos abaixo é o que geram a Figura:
plt.plot(tempo, FID_Real, tempo, FID_Imag, tempo, Modulo,'--', label="Real")
plt.fill_between(tempo, FID_Real, 195, where=(FID_Real > 195), facecolor='g', alpha=0.6)
plt.grid(True, linestyle='-.')
plt.legend(loc=1) # upper rigth corner
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (u.a.)')
plt.title("Sinal oscilatório: FID(t) = A * sen ($\omega$*t) * exp(-t/$T_{2}^{*}$)")
plt.show()
```

# Equações de Bloch

Vamos diretamente agora ver as equações de Bloch, após um pulso simples, a magnetização sofre dois processos
\begin{align}
         \left\{
        \begin{array}{cl}
        M_{x,y}(t) &=& M_{0} \mathrm{e}^{-t/T_{2}} \\
        M_{z}(t) &=& M_{0} (1- \mathrm{e}^{-t/T_{1}})
        \end{array}
        \right.
    \end{align}

```python
import numpy as np
from matplotlib import pyplot as plt

# Escolha o tempo de Relaxação Longitudinal (T2) 
T2 =  0.9

# eixo tempo
t = np.linspace(0, 10, 1000)

# M0
M0 = 1

# Equação Mxy(t) = M0 * exp (-t/T2)
Mxy = M0 * np.exp(-t/T2)

plt.plot(t, Mxy, '-', label=r"Decaimento $T_{2}$")
plt.fill_between(t, Mxy, 195, where=(Mxy > 195), facecolor='g', alpha=0.6)

plt.title("Decaimento Exponencial $T_{2}$")
plt.grid(True, linestyle='-.')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (u.a.)')
plt.legend(loc=1) # upper rigth corner
plt.show()
```

Vamos começar simulando basicamente as Equações de Bloch de modo direto:

```python
# Escolha o tempo de Relaxação Transversal (T1) 
T1 =  1.5

# Dica: Eixo do tempo já foi criado acima, não precisa fazer novamente.

# Equação Mz(t) = M0 * ( 1 - 2* exp(-t/T1) )
Mz = M0 * (1 - 2* np.exp(-t/T2))

plt.plot(t, Mz, '-', label=r"$T_{1}$")
plt.fill_between(t, Mz, 195, where=(Mz > 195), facecolor='g', alpha=0.6)
plt.title("Crescimento Exponencial $T_{1}$")
plt.grid(True, linestyle='-.')
plt.legend(loc=1) # upper rigth corner
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (u.a.)')
plt.show()
```

# Inomogeneidade de Campo Magnético

Campos magnéticos tem inomogeneidade

$$S(\omega) = \frac{T_{2}^{*}}{1+(\omega - \omega_{0})^{2}(T_{2}^{*})^{2}} $$

Com largura de linha a meia altura (*FWHM - Full width at half maximum*)

$$ FWHM = \frac{1}{\pi T_{2}^{*}}$$

```python
# Escolha o tempo T2* (ms)
T2estrela =  20

# Escolha a Frequência de ressonância  (em MHz)
freq_0 =  10

# largura distribuição (Hz)
FWHM = 1/(np.pi*T2estrela);

# Região redor pico (Isocromatas)
# 39*FWHM cobre 99,5% região isocromata
# 76*FWHM cobre 99,9% região isocromata
# Janela Espectral sempre centrada em freq_0 = 0! (Fref-F)
x01 = freq_0 - 20*FWHM/2; 
x02 = freq_0 + 20*FWHM/2;

# Offset de frequencia Omega = (2pi*gamma)*2pi*(fref - f0)
# df = Omega / 2pi
freq = np.linspace(x01, x02, 1000)

# Equação da Distribuição Lorentziana
def S(freq):
  global T2estrela
  global freq_0
  return T2estrela/ (1+ ((freq - freq_0)*2*np.pi*T2estrela)**2 )

S = S(freq)

plt.plot(freq, S, '-',label=r"Ressonância")
plt.fill_between(freq, S, 195, where=(S > 195), facecolor='g', alpha=0.6)
plt.grid(True, linestyle='-.')
plt.legend(loc=1) # upper rigth corner
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude (u.a.)')
plt.title("Distribuição Lorentziana")
plt.show()
```

Não existe campo magnético perfeito, mas caso fosse teriamos um único pico exatamente na Frequência de Ressonâincia (curva vermelha abaixo):

# Sequências de Pulsos

**Sequência de Pulsos** na RMN são projetadas para obter determinadas informações do sistema em estudo, por exemplo:

- **Seq. Pulso Simples**, obtém diretamente o sinal FID (*Free Induction Decay*) que contém informação da frequência de precessão dos spins ($\Omega$), amplitude do sinal é relacionada com a quantidade de spins, entre outras informações;

- **Seq. Spin-Echo**, com a  aplicação de dois pulsos (90° - $\tau$ - 180° - echo), obtemos um echo de spin, que pela sua amplitude é possivel extrair informação para quantificação de óleo em grãos, alimentos, entre diversas aplicações;

- **Seq. CPMG**, determinação da curva de relaxação $T_{2}$,informação que pode ser correlacionada com viscosidade, porosidade, entre muitas outras.

- **Seq. Inversão Recuperação (IR)**, determinação da curva de relaxação $T_{1}$, informação que pode ser correlacionada com viscosidade, porosidade, entre muitas outras.

Essas quatro sequências de pulsos vamos detalhar mais na próxima seção, e códigos para realizar simulações em função de vários fatores experimentais podem ser acessado aqui.
